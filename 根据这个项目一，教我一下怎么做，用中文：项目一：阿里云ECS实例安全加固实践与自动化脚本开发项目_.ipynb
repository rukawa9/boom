## 第一阶段：理论学习（第一周）

这个阶段是为你打下坚实基础。不要急于动手，充分理解每个安全措施的“为什么”能让你在后续实践中事半功倍。

### 1\. 阿里云ECS基础与安全最佳实践

  * **学习内容：** 理解 **ECS实例** 是什么（一台虚拟服务器），它的核心组成部分（CPU、内存、磁盘、网络），以及它在阿里云生态中的位置。
  * **学习资源：**
      * **阿里云官方文档：** 这是你最主要的学习资料。搜索“ECS产品文档”、“ECS安全最佳实践”或“安全组最佳实践”。
      * **重点关注：**
          * **ECS实例生命周期：** 如何创建、启动、停止、重启和删除实例。
          * **镜像：** 系统镜像（公共镜像、自定义镜像、云市场镜像）及其作用。
          * **网络：** VPC、VSwitch、弹性公网IP (EIP)、私网IP。
          * **存储：** 云盘 (ESSD、SSD、高效云盘)。
          * **监控：** 云监控。
          * **ECS安全最佳实践：** 寻找官方关于加固ECS实例的指导。

### 2\. Linux操作系统安全加固基础知识

大多数ECS实例都运行Linux。理解其安全方面至关重要。

  * **学习内容：**
      * **SSH安全：**
          * SSH是什么，它是如何实现远程访问的？
          * 为什么禁用root用户直接登录很重要？
          * 为什么修改默认SSH端口（22）？
          * **密钥登录与密码登录：** 了解密钥登录的优越性。如何生成SSH密钥（`ssh-keygen`命令）并管理它们。
          * `sshd_config`文件：学习常见参数，如`PermitRootLogin`、`Port`、`PasswordAuthentication`、`PubkeyAuthentication`、`AllowUsers`、`DenyUsers`。
      * **防火墙配置：**
          * 防火墙是什么？为什么实例上需要它？
          * **Ubuntu的`ufw` (Uncomplicated Firewall)：** 学习`ufw enable`、`ufw disable`、`ufw status`、`ufw allow/deny`等基本命令。
          * **CentOS/RHEL的`firewalld`：** 学习`firewall-cmd --state`、`firewall-cmd --zone=public --add-port`、`firewall-cmd --reload`等基本命令。
      * **用户与权限管理：**
          * `useradd`、`usermod`、`userdel`命令。
          * `passwd`命令用于设置密码。
          * `sudo`和`visudo`：如何在不直接使用root的情况下，授予用户管理员权限。
          * 文件权限（`chmod`、`chown`）及其重要性（`rwx` - 读取、写入、执行）。
      * **系统更新与补丁管理：**
          * 为什么定期更新对安全性至关重要。
          * `apt update`、`apt upgrade` (Ubuntu/Debian)。
          * `yum update` (CentOS/RHEL)。
  * **学习资源：**
      * 在线教程（例如，DigitalOcean、Linode上关于Linux安全的指南）。
      * Linux手册页（例如，`man sshd_config`、`man ufw`）。
      * 专注Linux安全加固的安全博客和文章。

## 第二阶段：实践操作（第二至第三周）

这是你亲自动手的阶段。创建一个专门的测试环境，以避免影响任何现有资源。

### 1\. 创建测试环境

1.  **登录阿里云控制台：** 使用你的阿里云账号登录。如果你使用的是子账号，请确保它拥有创建和管理ECS实例的必要权限。

2.  **进入ECS控制台：** 导航到“弹性计算服务”。

3.  **创建实例：**

      * 点击“创建实例”。
      * **地域与可用区：** 选择一个离你或项目目标最近的地域。
      * **实例类型：** 一个小型实例类型（例如`ecs.t5-lc1m1.small`或`ecs.t6-c1m1.large`）足以用于测试。
      * **镜像：** 选择**公共镜像**——**Ubuntu Server（最新的LTS版本，如22.04 LTS）或CentOS Stream 9（或7/8）**。Ubuntu通常对初学者更友好，因为它使用`ufw`。
      * **存储：** 默认系统盘即可。
      * **网络：**
          * **专有网络VPC：** 如果没有，可以创建一个新的VPC和VSwitch，或者使用一个现有的非生产环境VPC。
          * **公网IP：** 分配一个公网IP地址（如果需要它持久存在，建议选择EIP）。
          * **安全组：** 选择或**创建一个新的安全组**。最初，你可能需要允许从你自己的IP地址通过SSH（端口22）访问，以便于连接。*稍后你会对其进行精炼。*
      * **登录方式：** 选择**密钥对**（出于安全考虑强烈推荐）。如果你没有，点击“创建密钥对”，下载它并妥善保管（`.pem`文件）。
      * **确认：** 检查配置并创建实例。

4.  **连接到你的ECS实例：**

      * **对于Linux/macOS用户：** 打开终端。设置私钥的正确权限：`chmod 400 your_key_pair.pem`。然后连接：`ssh -i your_key_pair.pem username@your_instance_public_ip`（例如，`ssh -i mykey.pem root@123.45.67.89`）。
      * **对于Windows用户：** 可以使用**PuTTYgen**将`.pem`密钥转换为PuTTY使用的`.ppk`格式，或者使用\*\*Windows Subsystem for Linux (WSL)\*\*获得原生的`ssh`体验。

### 2\. 安全组配置

这是你网络层的第一道防线。

1.  **导航到实例的安全组：** 在ECS控制台中，找到你的实例，然后点击其关联的安全组ID。
2.  **入方向规则：**
      * **SSH（端口22）：** 最初，它可能对`0.0.0.0/0`（所有IP）开放。**立即更改此设置。** 编辑规则，仅允许从**你自己的公网IP地址**通过SSH访问（你可以在谷歌上搜索“what is my ip”找到你的公网IP）。
      * **其他必要端口：** 如果你的实例将运行Web服务（例如Nginx、Apache），添加\*\*HTTP（端口80）**和**HTTPS（端口443）\*\*的规则，同样地，如果可能，最好限制到特定的IP范围（例如，对于公共Web服务，可以设置为`0.0.0.0/0`，但请谨慎）。
      * **删除不必要的规则：** 移除任何过于宽松的默认规则。
3.  **出方向规则：** 通常，出方向规则限制较少（通常是`0.0.0.0/0`对所有端口），但在高安全环境中，你可以限制出站连接只允许必要的服务。对于本项目，你可以保持默认以简化操作，除非你想进一步探索。

### 3\. SSH安全加固（在ECS实例内部操作）

通过SSH连接到你的ECS实例。所有这些步骤都在你的Linux实例**内部**执行。

1.  **备份`sshd_config`文件：**
    ```bash
    sudo cp /etc/ssh/sshd_config /etc/ssh/sshd_config.bak
    ```
2.  **编辑`sshd_config`文件：**
    ```bash
    sudo nano /etc/ssh/sshd_config
    ```
    （如果你更喜欢，也可以用`vi`）
3.  **禁用root用户直接登录：**
      * 找到行：`#PermitRootLogin prohibit-password`（或`yes`）。
      * 修改为：`PermitRootLogin no`
      * *重要提示：* 在执行此操作之前，请确保你已经创建了一个具有`sudo`权限的非root用户（参见下面的“用户权限管理”），并且你可以使用该用户登录，否则你将把自己锁在外面！
4.  **更改默认SSH端口：**
      * 找到行：`#Port 22`
      * 修改为：`Port YourNewPortNumber`（例如，`Port 2222`或任何未被占用的1024以上的端口）。
      * *请记住：* 更改此设置后，你需要更新你的安全组规则以允许新端口，并且在连接时，指定`ssh -p YourNewPortNumber ...`。
5.  **使用密钥登录（禁用密码登录）：**
      * 确保你的密钥对登录已完美工作。
      * 找到行：`#PasswordAuthentication yes`
      * 修改为：`PasswordAuthentication no`
      * 找到行：`#PubkeyAuthentication yes`
      * 确保它已取消注释并设置为：`PubkeyAuthentication yes`
6.  **配置SSH登录失败次数限制（可选但建议）：**
      * 添加或修改以下行：
        ```
        MaxAuthAttempts 3  # 每次连接允许的最大认证尝试次数
        MaxStartups 5:30:100 # 最大未认证连接数：丢弃率：最大打开数（例如，同时最多5个，30%丢弃，100个最大）
        LoginGraceTime 60 # 登录完成的时间（秒）
        ```
7.  **保存并退出：** 保存`sshd_config`文件。
8.  **重启SSH服务：**
    ```bash
    sudo systemctl restart sshd  # 适用于Ubuntu/CentOS 7+
    # 或者对于旧版CentOS：sudo service sshd restart
    ```
9.  **测试新的SSH配置：** **切勿关闭你当前的SSH会话！** 打开一个**新的**终端窗口，尝试使用新端口和非root用户（如果你禁用了root登录）进行连接。如果新连接成功，你就可以关闭旧会话了。如果失败，你可以使用旧会话中的备份文件恢复更改。

### 4\. 防火墙配置（在ECS实例内部操作）

根据你的操作系统选择`ufw`（Ubuntu）或`firewalld`（CentOS）。

#### 对于Ubuntu（使用`ufw`）：

1.  **安装`ufw`（如果未安装）：**
    ```bash
    sudo apt update
    sudo apt install ufw
    ```
2.  **检查状态：** `sudo ufw status verbose`（应该显示为inactive）
3.  **设置默认策略：**
    ```bash
    sudo ufw default deny incoming  # 默认拒绝所有入站连接
    sudo ufw default allow outgoing  # 默认允许所有出站连接（之后可以限制）
    ```
4.  **允许SSH：**
    ```bash
    sudo ufw allow YourNewPortNumber/tcp  # 允许你的新SSH端口
    # 或者如果你没有修改，则为：sudo ufw allow 22/tcp
    ```
5.  **允许其他必要服务（例如HTTP/HTTPS）：**
    ```bash
    sudo ufw allow 80/tcp
    sudo ufw allow 443/tcp
    ```
6.  **启用防火墙：**
    ```bash
    sudo ufw enable
    ```
      * **确认：** `sudo ufw status verbose`应该显示为active并列出你允许的规则。

#### 对于CentOS（使用`firewalld`）：

1.  **检查状态：** `sudo systemctl status firewalld`（应该显示为running）
2.  **允许SSH：**
    ```bash
    sudo firewall-cmd --permanent --add-port=YourNewPortNumber/tcp
    # 或者如果你没有修改，则为：sudo firewall-cmd --permanent --add-service=ssh
    ```
3.  **允许其他必要服务（例如HTTP/HTTPS）：**
    ```bash
    sudo firewall-cmd --permanent --add-service=http
    sudo firewall-cmd --permanent --add-service=https
    ```
4.  **重新加载防火墙规则：**
    ```bash
    sudo firewall-cmd --reload
    ```
      * **确认：** `sudo firewall-cmd --list-all`应该显示你的活动规则。

### 5\. 用户权限管理（在ECS实例内部操作）

1.  **创建新的非root用户：**
    ```bash
    sudo adduser your_new_username
    ```
    按照提示设置密码和用户信息。
2.  **将新用户添加到`sudo`组（用于管理权限）：**
    ```bash
    sudo usermod -aG sudo your_new_username  # 适用于Ubuntu/Debian
    # 或者适用于CentOS/RHEL：sudo usermod -aG wheel your_new_username
    ```
3.  **为新用户复制SSH公钥：**
    如果你想使用SSH密钥登录此新用户，请将你的公钥内容（`.pub`文件的内容）复制到新用户的`~/.ssh/authorized_keys`文件中。
    ```bash
    # 在root或其他具有sudo权限的用户登录时
    sudo su - your_new_username
    mkdir -p ~/.ssh # -p 确保父目录也创建
    chmod 700 ~/.ssh
    nano ~/.ssh/authorized_keys
    # 在这里粘贴你的公钥内容
    chmod 600 ~/.ssh/authorized_keys
    exit # 从新用户的shell退出
    ```
4.  **测试以新用户登录：** 打开一个新的终端，尝试`ssh -i your_key_pair.pem your_new_username@your_instance_public_ip -p YourNewPortNumber`。

### 6\. 系统更新与补丁管理（在ECS实例内部操作）

这对于修补已知漏洞至关重要。

  * **对于Ubuntu：**
    ```bash
    sudo apt update   # 刷新可用软件包列表
    sudo apt upgrade  # 升级所有已安装的软件包到最新版本
    sudo apt autoremove # 删除不再需要的软件包
    ```
  * **对于CentOS：**
    ```bash
    sudo yum update   # 更新所有已安装的软件包
    ```
    在进行重要的内核更新后，通常建议重启系统：`sudo reboot`。

## 第三阶段：自动化脚本开发（第四周）

这是激动人心的部分，你将把手动操作转化为可重用的脚本。专注于核心的加固步骤。

### 1\. 选择你的脚本语言

  * **Shell脚本 (Bash)：** 非常适合系统级任务、编辑配置文件、运行命令。对于Linux系统加固来说，它通常是最快的选择。
  * **Python：** 功能更强大，适用于复杂的逻辑、错误处理，尤其适用于与**阿里云SDK**交互（例如，用于安全组自动化）。

**建议：** 对于实例内部的加固（SSH、防火墙、用户），先从**Bash**开始。如果你想自动化安全组配置，那么**Python结合阿里云Python SDK**是更好的选择。

### 2\. 实例内部加固脚本（Bash示例）

让我们创建一个基本的Bash脚本，用于自动化一些`sshd_config`的更改和`ufw`的设置。

  * **创建文件：** `nano harden_ecs.sh`

  * **内容：**

    ```bash
    #!/bin/bash

    # --- 配置变量 ---
    NEW_SSH_PORT="2222"             # 你希望的新SSH端口
    NEW_USER="intern_admin"         # 要创建的新sudo用户
    # 注意：请替换为你的实际公钥内容（来自 ~/.ssh/id_rsa.pub）。
    # 这个公钥将添加到新创建用户的authorized_keys中，用于SSH登录。
    YOUR_PUBLIC_KEY="ssh-rsa AAAAB3NzaC1yc2EAAAADAQABAAQC...你的完整公钥内容..."

    # --- 检查root权限 ---
    if [ "$EUID" -ne 0 ]; then
      echo "请以root用户或使用sudo运行此脚本。"
      exit 1
    fi

    echo "--- 正在启动ECS实例加固脚本 ---"

    # --- 1. 系统更新 ---
    echo "正在更新系统软件包..."
    apt update && apt upgrade -y
    apt autoremove -y
    echo "系统更新完成。"

    # --- 2. SSH安全加固 ---
    echo "正在加固SSH配置..."
    SSH_CONFIG="/etc/ssh/sshd_config"
    cp $SSH_CONFIG "${SSH_CONFIG}.bak" # 备份原始文件

    # 更改SSH端口
    sed -i "s/^#Port 22/Port $NEW_SSH_PORT/" $SSH_CONFIG
    sed -i "s/^Port 22/Port $NEW_SSH_PORT/" $SSH_CONFIG # 以防它已取消注释

    # 禁用Root登录
    sed -i "s/^PermitRootLogin yes/PermitRootLogin no/" $SSH_CONFIG
    sed -i "s/^#PermitRootLogin prohibit-password/PermitRootLogin no/" $SSH_CONFIG # 适用于较新系统

    # 禁用密码认证
    sed -i "s/^PasswordAuthentication yes/PasswordAuthentication no/" $SSH_CONFIG
    sed -i "s/^#PasswordAuthentication no/PasswordAuthentication no/" $SSH_CONFIG # 确保明确设置为no

    # 启用公钥认证（确保已启用）
    sed -i "s/^#PubkeyAuthentication yes/PubkeyAuthentication yes/" $SSH_CONFIG

    # 添加登录失败限制（如果不存在）
    grep -q "MaxAuthAttempts" $SSH_CONFIG || echo "MaxAuthAttempts 3" >> $SSH_CONFIG
    grep -q "MaxStartups" $SSH_CONFIG || echo "MaxStartups 5:30:100" >> $SSH_CONFIG
    grep -q "LoginGraceTime" $SSH_CONFIG || echo "LoginGraceTime 60" >> $SSH_CONFIG

    echo "重启SSH服务..."
    systemctl restart sshd
    echo "SSH配置已更新。请记住测试新端口：ssh -p $NEW_SSH_PORT ..."

    # --- 3. 防火墙配置 (UFW) ---
    echo "正在配置UFW防火墙..."
    apt install ufw -y
    ufw default deny incoming
    ufw default allow outgoing
    ufw allow $NEW_SSH_PORT/tcp
    ufw allow 80/tcp    # 示例：允许HTTP
    ufw allow 443/tcp   # 示例：允许HTTPS
    ufw enable <<<$'y\n' # 自动应答提示
    echo "UFW防火墙已配置并启用。"

    # --- 4. 用户创建和Sudo权限 ---
    echo "正在创建新的sudo用户：$NEW_USER"
    if id "$NEW_USER" &>/dev/null; then
        echo "用户 $NEW_USER 已存在，跳过创建。"
    else
        adduser --disabled-password --gecos "" $NEW_USER # 创建用户但不设置密码
        usermod -aG sudo $NEW_USER # 将用户添加到sudo组
        echo "$NEW_USER ALL=(ALL) NOPASSWD: ALL" >> /etc/sudoers.d/$NEW_USER # 允许此用户无密码使用sudo
        chmod 440 /etc/sudoers.d/$NEW_USER

        # 配置新用户的SSH密钥登录
        mkdir -p /home/$NEW_USER/.ssh
        echo "$YOUR_PUBLIC_KEY" > /home/$NEW_USER/.ssh/authorized_keys
        chown -R $NEW_USER:$NEW_USER /home/$NEW_USER/.ssh
        chmod 700 /home/$NEW_USER/.ssh
        chmod 600 /home/$NEW_USER/.ssh/authorized_keys
        echo "用户 $NEW_USER 已创建，并配置了SSH密钥登录。"
    fi

    echo "--- ECS实例加固脚本执行完毕！ ---"
    echo "请尝试使用新用户和新SSH端口进行登录。"
    ```

  * **保存脚本。**

  * **添加执行权限：** `chmod +x harden_ecs.sh`

  * **运行脚本：** `sudo ./harden_ecs.sh`

      * **重要提示：** 在运行脚本之前，请再次确认你的`YOUR_PUBLIC_KEY`变量已正确替换为你的**实际公钥内容**。运行前，请确保在测试环境中进行，避免影响生产环境。

### 3\. 自动化配置安全组（Python + Alibaba Cloud SDK示例）

这个部分需要Python环境和安装阿里云SDK。

1.  **安装Python和pip：**
    ```bash
    sudo apt install python3 python3-pip -y # Ubuntu
    # 或者 sudo yum install python3 python3-pip -y # CentOS
    ```
2.  **安装阿里云Python SDK：**
    ```bash
    pip3 install aliyun-python-sdk-core
    ```

  * **内容：**

````
#!/usr/bin/env python3
    # -*- coding: utf-8 -*-

    from aliyunsdkcore.client import AcsClient
    from aliyunsdkcore.acs_exception.exceptions import ClientException, ServerException
    from aliyunsdkecs.request.v20140526 import AuthorizeSecurityGroupRequest
    from aliyunsdkecs.request.v20140526 import RevokeSecurityGroupRequest
    import os

    # --- 配置信息 ---
    # 从环境变量或凭证文件中读取，这里直接写，实际项目中建议从环境变量读取
    # ACCESS_KEY_ID = os.environ.get('ALIYUN_ACCESS_KEY_ID')
    # ACCESS_KEY_SECRET = os.environ.get('ALIYUN_ACCESS_KEY_SECRET')
    # REGION_ID = os.environ.get('ALIYUN_REGION_ID')

    # 为了简化演示，这里直接填写，实际请确保安全，不要硬编码在代码中
    # 请替换为你的实际信息
    ACCESS_KEY_ID = 'YOUR_ALIYUN_ACCESS_KEY_ID'
    ACCESS_KEY_SECRET = 'YOUR_ALIYUN_ACCESS_KEY_SECRET'
    REGION_ID = 'cn-hangzhou' # 替换为你的ECS实例所在区域，例如 'ap-southeast-1' (新加坡), 'us-west-1' (美国西部)

    SECURITY_GROUP_ID = 'sg-your_security_group_id' # 替换为你的安全组ID
    YOUR_PUBLIC_IP = 'your.public.ip.address/32' # 替换为你的公网IP地址，带上/32表示精确匹配

    NEW_SSH_PORT = '2222' # 对应你脚本中修改的SSH端口

    def get_client():
        return AcsClient(ACCESS_KEY_ID, ACCESS_KEY_SECRET, REGION_ID)

    def authorize_security_group_rule(client, security_group_id, port_range, ip_protocol, source_cidr_ip, policy='accept', nic_type='internet'):
        request = AuthorizeSecurityGroupRequest.AuthorizeSecurityGroupRequest()
        request.set_SecurityGroupId(security_group_id)
        request.set_PortRange(port_range)
        request.set_IpProtocol(ip_protocol)
        request.set_SourceCidrIp(source_cidr_ip)
        request.set_Policy(policy)
        request.set_NicType(nic_type) # internet 或 intranet

        try:
            response = client.do_action_with_exception(request)
            print(f"Rule added: Port {port_range}, Protocol {ip_protocol}, Source {source_cidr_ip}")
            return True
        except ClientException as e:
            if "RuleAlreadyExists" in str(e):
                print(f"Rule already exists: Port {port_range}, Protocol {ip_protocol}, Source {source_cidr_ip}")
                return True
            else:
                print(f"Failed to add rule: {e}")
                return False
        except ServerException as e:
            print(f"Server error when adding rule: {e}")
            return False

    def revoke_security_group_rule(client, security_group_id, port_range, ip_protocol, source_cidr_ip, policy='accept', nic_type='internet'):
        request = RevokeSecurityGroupRequest.RevokeSecurityGroupRequest()
        request.set_SecurityGroupId(security_group_id)
        request.set_PortRange(port_range)
        request.set_IpProtocol(ip_protocol)
        request.set_SourceCidrIp(source_cidr_ip)
        request.set_Policy(policy)
        request.set_NicType(nic_type) # internet 或 intranet

        try:
            response = client.do_action_with_exception(request)
            print(f"Rule revoked: Port {port_range}, Protocol {ip_protocol}, Source {source_cidr_ip}")
            return True
        except ClientException as e:
            if "InvalidRule.NotFound" in str(e):
                print(f"Rule not found: Port {port_range}, Protocol {ip_protocol}, Source {source_cidr_ip}")
                return True
            else:
                print(f"Failed to revoke rule: {e}")
                return False
        except ServerException as e:
            print(f"Server error when revoking rule: {e}")
            return False

    if __name__ == "__main__":
        client = get_client()

        print("--- Updating Security Group Rules ---")

        # 1. 删除旧的SSH端口规则 (如果存在，并且是针对你的IP)
        # 假设你之前是开放22端口给你的IP
        revoke_security_group_rule(client, SECURITY_GROUP_ID, '22/22', 'tcp', YOUR_PUBLIC_IP)

        # 2. 添加新的SSH端口规则
        authorize_security_group_rule(client, SECURITY_GROUP_ID, f'{NEW_SSH_PORT}/{NEW_SSH_PORT}', 'tcp', YOUR_PUBLIC_IP)

        # 3. 添加HTTP和HTTPS规则（如果需要对外开放）
        # 这里为了演示，我们假设你的Web服务是对所有公网开放的
        authorize_security_group_rule(client, SECURITY_GROUP_ID, '80/80', 'tcp', '0.0.0.0/0')
        authorize_security_group_rule(client, SECURITY_GROUP_ID, '443/443', 'tcp', '0.0.0.0/0')

        # 4. 删除其他可能过于开放的规则，例如22端口对0.0.0.0/0开放的规则
        # 这需要你根据实际情况判断和操作，非常危险，请务必小心！
        # revoke_security_group_rule(client, SECURITY_GROUP_ID, '22/22', 'tcp', '0.0.0.0/0')
        # 仅在你确认安全组中存在这条并且你想删除它时才取消注释并运行！

        print("--- Security Group Rules Update Complete ---")
````

          * **`YOUR_PUBLIC_IP`** 必须是**带CIDR后缀的**，例如`1.2.3.4/32`。
